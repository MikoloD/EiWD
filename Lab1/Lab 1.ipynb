Zadanie 40  przedstawic sposoby laczenia ramek danych za pomoca˛ metod merge i concat

```python
df1 = pd.read_csv("IHME-GBD_2019_DATA-15798851-2.csv", encoding='utf-8', engine='python')
df2 = pd.read_csv("IHME-GBD_2019_DATA-ff08d9bc-1.csv", encoding='utf-8', engine='python')
```

```python
df1.head()
```

```python
df2.head()
```

```python
df1.rename(columns = {'val': 'val1', 'upper':'upper1', 'lower':'lower1'}, inplace = True)
df2.rename(columns = {'val': 'val2', 'upper': 'upper2', 'lower':'lower2'}, inplace = True)
df_both = pd.merge(df1, df2, on = ['location', 'sex', 'age', 'cause', 'metric', 'year'], how = 'inner') 
df_both.head()
```

```python
#concat
df_part1 = df_both.iloc[:50000,:]
df_part2 = df_both.iloc[50000:,:]
df_both_2 = pd.concat([df_part1, df_part2], axis = 0)
df_both_2.head()
```

Zadanie 41 pokazac dodawanie nowych kolumn za pomoca operacji matematycznych

```python
df_both["val2_round"] = df_both["val2"].round(decimals = 1) 
df_both.head()
```

```python
df_both["sum"] = df_both["val2"] + df_both["upper2"] + df_both["upper1"]
df_both.head()
```

```python
df_both["sum2"] = df_both.loc[:, "val2":"lower2"].sum(axis = 1)
df_both.head()
```

Zadanie 42 przedstawic na przykladzie dodawanie nowych kolumn z pomoca funkcji lambda

```python
c = ['Poland', 'Ukraine', 'Italy', 'Germany', 'Hungary', 'Spain', 'Romania']
```

```python
df_both['c'] = df_both['location'].apply(lambda x: True if x in c else False )
df_both[df_both['c'] == True]
```

Zadanie 43 przedstawic mozliwosci pracy z duzymi plikami przy uzyciu argumentu chunksize

```python
df_both.to_csv('df_both.csv')

for chunk_df in pd.read_csv('df_both.csv', 
                      chunksize = 50000):
    print("CHUNK DF")
    print(chunk_df.tail())
```
